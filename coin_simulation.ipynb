```python
import random
import matplotlib.pyplot as plt
```

```python
stored_outcome = ['head','tail']
storage = []

for i in range(1,100):
    outcome = random.sample(stored_outcome,1)
    storage.append(outcome[0])

print(f'head = {storage.count("head")} and tail = {storage.count("tail")}')

categories = ['head','tail']
head = storage.count("head")
tail = storage.count("tail")

plt.bar(categories, [head,tail], color=['red','blue'])
plt.title('Heads and Tails')
plt.show()
```
